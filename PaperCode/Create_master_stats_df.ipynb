```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#open stats data

mypath = '/Volumes/Seagate-stars/PAPER_FINAL_FILES/stats/'
#ls
ls_stitched = pd.read_csv('{}stitched_ls_statsdf.csv'.format(mypath))
ls_sectors1 = pd.read_csv('{}ls_statsdf.csv'.format(mypath))
ls_sectors2 = pd.read_csv('{}ls_statsdf_14.csv'.format(mypath))#add this in b/c of path error sec14 got skipped
ls_sectors = pd.concat([ls_sectors1,ls_sectors2]).reset_index(drop=True)
ls_sectors = ls_sectors.drop_duplicates(subset=['TIC','Sector']) #b/c over 1415/ncvz overlap targets done twice
#bls
bls_sectors = pd.read_csv('{}bls_stats_sectors.csv'.format(mypath))
bls_stitched = pd.read_csv('{}bls_stats_stitched.csv'.format(mypath))
print('ls stitched: {}, ls sectors: {}'.format(len(ls_stitched),len(ls_sectors)))
print('bls stitched: {}, bls sectors: {}'.format(len(bls_stitched),len(bls_sectors)))
print('unique ls: stitched = {}, sectors= {}'.format(len(np.unique(ls_stitched['TIC'].to_numpy())), len(np.unique(ls_sectors['TIC'].to_numpy()))))
print('unique bls: stitched = {}, sectors= {}'.format(len(np.unique(bls_stitched['TIC'].to_numpy())), len(np.unique(bls_sectors['TIC'].to_numpy()))))


#merge sectors with stitched for each analysis
ls = pd.concat([ls_sectors,ls_stitched]).reset_index(drop=True)
bls = pd.concat([bls_sectors,bls_stitched]).reset_index(drop=True)

print('\nls total: {}, bls total: {}'.format(len(ls),len(bls)))
print('unique totals: ls={}, bls={}'.format(len(np.unique(ls['TIC'].to_numpy(str))), len(np.unique(bls['TIC'].to_numpy(str)))))
```

```python
#see which targets were dropped in merge with ticcat

#ticcat used the new 2021 updated target lists for sectors 1-26
#ls/bls stats analyzed the target lists for sectors 1-26 available in 2020
#this is why expected some to not match up
#this cell seeks to find which weren't analyzed yet & which were dropped from 2020 to 2021
def returnNotMatches(a, b):
    '''
    inputs: a,b
            both lists or arrays
    outputs: not in b, not in a
            two lists
    '''
    return [[x for x in a if x not in b], [x for x in b if x not in a]] #not in b , not in a


#for number of light curves
a = ticcat['ID'].to_numpy()
b = stats_df['TIC'].to_numpy()
notinb, notina = returnNotMatches(a,b)
print(len(notina), len(notinb))
#for number of unique targets
a = np.unique(ticcat['ID'].to_numpy())
b = np.unique(stats_df['TIC'].to_numpy())
notinb, notina = returnNotMatches(a,b)
print(len(notina), len(notinb),notina)

```
